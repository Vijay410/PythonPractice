# Python Operators

1. Python operators are symbols that perform operations on variables and values. They are categorized into several types:

There are several types of operators:
1. Arithmetic Operators
2. Comparison (Relational) Operators
3. Logical Operators
4. Bitwise Operators
5. Assignment Operators
6. Identity Operators
7. Membership Operators


1. Airthmetic Operator:




```python
#Basics Airthmetic Operations
a = 15
b = 4

print("Addition:", a + b)        
print("Subtraction:", a - b)     
print("Multiplication:", a * b)  
print("Division:", a / b)        
print("Floor Division:", a // b) # return whole number
print("Modulus:", a % b)         
print("Exponentiation:", a ** b) 
```

#### 2. Comparison (Relational) Operators

 


```python
# Lets see some examples on comparision operators
# If the condition is satisfied which will return True / False
a = 10
b = 20

print(f"a is greater than b? : ", a > b)
print(f"a is less than b? : ", a < b)
print(f"a is equal than b? : ", a == b)
print(f"a is >= than b? : ", a >= b)
print(f"a is <= than b? : ", a <= b)
print(f"a is != than b? : ", a != b)
```

## 3. Logical Operators



```python
#  Lets understand Logical operators with examples
a = 10
b = 5
c = 20

print(a > b and c > a)  # True (both conditions are True)
print(a > c and b < a)  # False (one condition is False)
print(a > c or b < a)  # False (one condition is False)
print(a is not b) # Check a is same as b,
print(a is b)
```

#### 4. Bitwise Operators


```python
#And operation
a = 5  #  0101
b = 3  #  0011
result = a & b  # 0001 (1 in decimal)
print(result)  # Output: 1

#OR Operation
a = 5  #  0101
b = 3  #  0011
result = a | b  
print(result)

#XOR Operation
a = 5  #  0101
b = 3  #  0011
result = a ^ b
print(result)  

#Right Shift
a = 5  #  0101
b = 3  #  0011
result = a >> b
print(result)

#Left shift
a = 5  #  0101
b = 3  #  0011
result = a << b
print(result)  

a = 5  #  0101
result = ~a  # -(5+1) = -6
print(result)  # Output: -6
```

#### 5. Assignment Operators


```python
# Simple assignment
a = 10  
print(a)  # Output: 10

# Addition assignment
a += 5   # Equivalent to: a = a + 5
print(a)  # Output: 15

# Subtraction assignment
a -= 3   # Equivalent to: a = a - 3
print(a)  # Output: 12

# Multiplication assignment
a *= 2   # Equivalent to: a = a * 2
print(a)  # Output: 24

# Division assignment
a /= 4   # Equivalent to: a = a / 4
print(a)  # Output: 6.0

# Floor division assignment
a //= 2  # Equivalent to: a = a // 2
print(a)  # Output: 3.0

# Modulus assignment
a %= 2   # Equivalent to: a = a % 2
print(a)  # Output: 1.0

```

### 7. Membership Operators


```python
# Using 'in' with lists
fruits = ["apple", "banana", "cherry"]
print("apple" in fruits)   # Output: True
print("grape" in fruits)   # Output: False

# Using 'not in' with lists
print("grape" not in fruits)  # Output: True

# Using 'in' with strings
text = "Hello, World!"
print("H" in text)  # Output: True
print("hello" in text)  # Output: False (case-sensitive)

# Using 'not in' with strings
print("Python" not in text)  # Output: True

# Using 'in' with tuples
numbers = (1, 2, 3, 4, 5)
print(3 in numbers)   # Output: True
print(10 in numbers)  # Output: False

# Using 'not in' with sets
unique_numbers = {10, 20, 30, 40}
print(20 in unique_numbers)   # Output: True
print(50 not in unique_numbers)  # Output: True
```
